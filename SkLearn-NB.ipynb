```python
import requests,csv
data = requests.get("https://s3.amazonaws.com/dataviz-curriculum/day_2/yelp_reviews.csv").content.decode('utf8')
```

```python
data[:50]
```

```python
data = data.split("\r")
data[:10]
```

```python
wdf = pandas.DataFrame([f for f in csv.reader(data)][1:], columns=['class','text'])
wdf
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(wdf.text)
X_train_counts.shape
```

```python
count_vect.vocabulary_.get(u'algorithm')
```

```python
count_vect.vocabulary_
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)
X_train_tf = tf_transformer.transform(X_train_counts)
X_train_tf.shape
```

```python
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train_tfidf, wdf['class'])
```

```python
docs_new = ['God is love', 'OpenGL on the GPU is fast', 'Im tired', 'I wish I was eating','Im sick', "It was terrible"]
X_new_counts = count_vect.transform(docs_new)
X_new_tfidf = tfidf_transformer.transform(X_new_counts)

predicted = clf.predict(X_new_tfidf)

for doc, category in zip(docs_new, predicted):
    print('%r => %s' % (doc, category))
```

```python
vectorizer_file = "tokenizer.sklearn"
pickle.dump(count_vect, open(tokenizer_file,'wb'))

tokenizer_file = "vectorizer.sklearn"
pickle.dump(tfidf_transformer, open(tokenizer_file,'wb'))

NBModel = 'simple_sentiment_scoring.sklearn'
pickle.dump(clf, open(NBModel, 'wb'))
```

```python
vectorizer = pickle.load(open(vectorizer_file, 'rb'))
tokenizer = pickle.load(open(tokenizer_file, 'rb'))
nbModel = pickle.load(open(NBModel, 'rb'))
```

```python
def predictSentiment(phrases: list):
    """
    Pass a list of one or more phrases, get back their sentiment.
    """
    X = vectorizer.transform(phrases)
    X = tokenizer.transform(X)
    y_predicted = nbModel.predict(X)
    return dict(zip(phrases, y_predicted))
```

```python
%%timeit -n 5
predictSentiment(['Own your agency', 'Its nice out', 'Im happy'])
```
